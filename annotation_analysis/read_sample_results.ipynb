### Read the manual annotated samples

```python
from pathlib import Path
import pandas as pd
```

```python
pwd = Path.cwd()
output_dir = pwd / 'samples'

sampleA_df = pd.read_csv(output_dir / 'sampled_A.csv')
sampleB_df = pd.read_csv(output_dir / 'sampled_B.csv')

samples_df = pd.concat([sampleA_df, sampleB_df], ignore_index=True)
```

```python
counts = (
    samples_df
    .groupby(["dataset", "label_correctness"])
    .size()
    .unstack(fill_value=0)
)

counts["total"] = counts.sum(axis=1)

print(counts.to_markdown())
print()
print(counts.to_latex())

# Output:
# 0 for incorrect, 1 for correct
# 2 for unclear

 
```
