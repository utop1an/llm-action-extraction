```python
filepath = os.path.join(pwd, "../../results", "cooking_nl2p_gpt-4o-mini.pkl")
indata = load_pkl(filepath)
for item in indata:
    words = item['words']
    acts = item['acts']
    pred = item['pred']

    act_names = []
    for act in acts:
        act_idx = act['act_idx']
        act_name = words[act_idx]
        act_names.append(act_name)
    pred_act_names = []
    for pred_act in pred:
        pred_act_names.append(pred_act['verb'])
    print("GT acts: ", act_names)
    print("Pred acts: ", pred_act_names)
    print()
    
```
