```python
import spacy
nlp = spacy.load("en_core_web_sm")


def match_obj(gt_name, pred_name):
    gt_lemma = nlp(gt_name)[0].lemma_.lower()
    doc2 = nlp(pred_name)
    pred_lemma = " ".join([token.lemma_.lower() for token in doc2])
    print("comparing gt_lemma=%s with pred_lemma=%s" % (gt_lemma, pred_lemma))
    return gt_lemma in pred_lemma
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
domains = ["cooking", "wikihow", "win2k"]

CAUSATIVE_VERBS = {"let", "allow", "have"}

def has_causative(sent):
    doc = nlp(sent)
    for token in doc:
        if token.lemma_ in {"keep", "maintain"}:
            if any(c.dep_ in {"xcomp", "acomp"} for c in token.children):
                return True
    return False

res = []
for domain in domains:
    filepath = os.path.join(pwd, "../../data", f"easdrl/{domain}_labeled_text_data.pkl")
    indata = load_pkl(filepath)
    count = 0
    total=0
    for item in indata:
        sents = item['sents']
        total += len(sents)
        for sent in sents:
            sent_text = " ".join(sent)
            if has_causative(sent_text):
                print(f"[{domain}] Causative found: {sent_text}")
                count += 1
    res.append((domain, count, total))

for domain, count, total in res:
    print(f"Domain: {domain}, Causative sentences: {count}/{total}")
```

```python
domains = ["cooking", "wikihow", "win2k"]
models = ["gpt-4o-mini", "gemma3_12b", "gpt-4.1-mini","gpt-4.1", "gpt-4o"]
PRONOUNS = {"I", "you", "he", "she", "it", "we", "they", "me", "him", "her", "us", "them"}

def has_pronoun(string):
    doc = nlp(string)
    for token in doc:
        if token.text in PRONOUNS:
            return True
    return False


res = []
for model in models:
    for domain in domains:
        filepath = os.path.join(pwd, "../../results/nl2p_1", f"{model}/{domain}_nl2p_1_{model}.pkl")
        indata = load_pkl(filepath)
        count = 0
        total=0
        for item in indata:
            preds = item.get('pred') or []
            for pred in preds:
                args = pred.get('arguments', [])
      
                total += len(args)
                for arg in args:
                    
                    if has_pronoun(arg):
                        print(f"[{domain}] Causative found: {arg}")
                        count += 1
        res.append((model, domain, count, total))

for model, domain, count, total in res:
    print(f"Model: {model}, Domain: {domain}, Causative sentences: {count}/{total}")
```
